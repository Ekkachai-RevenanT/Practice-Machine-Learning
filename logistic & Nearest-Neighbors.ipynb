```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt

```

```python
#Logistic(min-max)
```

```python
df = pd.read_csv('./data/df_final.csv')

X = df[df.columns[:-1]]
y = df['HeartDisease']

display(X.head())
display(y.head())

X_train, X_test, y_train, y_test = train_test_split(  X, y, test_size = 0.20, random_state = 42)

print('X_train (row/columns):',X_train.shape)
print('y_train (row):',y_train.shape)
print('X_test(row/columns):',X_test.shape)
print('y_test (row):',y_test.shape)

classifier =  LogisticRegression()
classifier.fit(X_train,y_train)

y_predict = classifier.predict(X_test)
y_score = classifier.predict_proba(X_test)[:,1]

print("Model Accuracy: ",accuracy_score(y_test,y_predict))

cm = confusion_matrix(y_test, y_predict)
ConfusionMatrixDisplay(cm).plot()
```

```python
df = pd.read_csv('./data/df_finish.csv')

X = df[df.columns[:-1]]
y = df['HeartDisease']

display(X.head())
display(y.head())

X_train, X_test, y_train, y_test = train_test_split(  X, y, test_size = 0.20, random_state = 42)

print('X_train (row/columns):',X_train.shape)
print('y_train (row):',y_train.shape)
print('X_test(row/columns):',X_test.shape)
print('y_test (row):',y_test.shape)

classifier =  LogisticRegression()
classifier.fit(X_train,y_train)

y_predict = classifier.predict(X_test)
y_score = classifier.predict_proba(X_test)[:,1]

print("Model Accuracy: ",accuracy_score(y_test,y_predict))

cm = confusion_matrix(y_test, y_predict)
ConfusionMatrixDisplay(cm).plot()
```
